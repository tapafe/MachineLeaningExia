```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
import scipy
import pandas as pd
import os
from scipy import misc
from skimage.io import imread_collection, imsave
from skimage import *
import cv2
from skimage.restoration import estimate_sigma

#creating a collection with the available images
noisy = imread_collection('dataset/Noisy/*.jpg')
blurry = imread_collection('dataset/Blurry/*.jpg')
```

```python
noisyTreated = imread_collection('dataset/noisyEnd/*.jpg')

def noise_difference(collection1, collection2):
    Df= pd.DataFrame(columns=['FileName','Sigma before','Sigma after'])
    for file in range(len(collection1)):
        ImageO = estimate_sigma(collection1[file], multichannel=True, average_sigmas=True) 
        ImageT = estimate_sigma(collection2[file], multichannel=True, average_sigmas=True)
        Df=Df.append({'FileName':collection1.files[file],'Sigma before':ImageO,'Sigma after':ImageT, 'Difference':ImageT-ImageO},ignore_index=True)
    performance_index_noisy = (Df["Sigma after"].mean() - Df["Sigma before"].mean())/Df["Sigma before"].mean()*100
    print("Différence max : ","\n", Df[ Df['Difference'] == Df['Difference'].max() ],"\n")
    print("Différence min : ","\n", Df[ Df['Difference'] == Df['Difference'].min() ],"\n")
    print("Moyenne sigma avant : ",Df["Sigma before"].mean(), "  |  ", "Moyenne sigma après : ", Df["Sigma after"].mean(), "\n")
    print("Pourcentage de différence : ", round(performance_index_noisy,2),"%","\n")
    print("Variance après traitement : ",Df["Sigma after"].var(), "\n")
    return Df
```

```python
noise_difference(noisy,noisyTreated)
```

```python
Interprétation des valeurs obtenues :
    
La fonction de traitement nous permets d'interpréter plusieurs informations. Tout d'abord, cela nous montre que l'image 14 de notre dataset est celle qui a subit le traitement 
de débruitage le plus important, et que la 71 est celle qui a subit le moins de traitement. En effet, plus la valeur de sigma est grande après traitement, plus l'image a été
débruitée. Ici, on obtient respectivement un sigma variant de 18.77 à 35.77 (soit une différence de 17) pour l'image 14, et une baisse de 5.5 à 2.5 soit une différence d'environ 3.
De plus, nous pouvons voir que la moyenne des sigma avant traitement est de 3.12 alors qu'elle est de 4.37 environ après traitement, sois une augmentation d'environ 40%, ce qui
prouve que notre algorithme a débruité toutes les images d'environ 40%, et donc qu'il est performant. Les valeurs obtenues nous permettent de déduire que la majorité des images on 
été débruitées, mais qu'une petite partie de celle-ci on potentiellement "perdu" de l'information.
La variance obtenue de 26 nous montre que notre traitement n'est pas optimal car elle n'est pas proche de 0. En effet, cela s'explique par les différences des sigmas obtenues sur
les différentes images après traitement.
```
