```python
from scipy import misc
import imageio
import matplotlib.pyplot as plt
import numpy as np
import scipy
import os
from skimage.io import imread_collection
from scipy import ndimage


#image path 
noisy_dir = 'Dataset/Noisy/*.jpg'

#creating a collection with the available images
noisy = imread_collection(noisy_dir)
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
#application of the Gaussian filter
def GaussianTreatment(image,sigma):
    
    k = np.array([[0,0,0],[0,1,0],[0,0,0]])
    
    red = image[:, :, 0]
    gauss_denoised_red = ndimage.gaussian_filter(red, sigma)
    fred = scipy.ndimage.convolve(gauss_denoised_red, k, mode='constant', cval=0.0)

    green = image[:, :, 1]
    gauss_denoised_green = ndimage.gaussian_filter(green, sigma)
    fgreen = scipy.ndimage.convolve(gauss_denoised_green, k, mode='constant', cval=0.0)

    blue = image[:, :, 2]
    gauss_denoised_blue = ndimage.gaussian_filter(blue, sigma)
    fblue = scipy.ndimage.convolve(gauss_denoised_blue, k, mode='constant', cval=0.0)
    
    return np.dstack((fred,fgreen,fblue))
#median filter function 
def medianTreatment(image,size):
    
    k = np.array([[0,0,0],[0,1,0],[0,0,0]])
    
    red = image[:, :, 0]
    median_denoised_red = ndimage.median_filter(red, size)
    fred = scipy.ndimage.convolve(median_denoised_red, k, mode='constant', cval=0.0)

    green = image[:, :, 1]
    median_denoised_green = ndimage.median_filter(green, size)
    fgreen = scipy.ndimage.convolve(median_denoised_green, k, mode='constant', cval=0.0)

    blue = image[:, :, 2]
    median_denoised_blue = ndimage.median_filter(blue, size)
    fblue = scipy.ndimage.convolve(median_denoised_blue, k, mode='constant', cval=0.0)
    
    return np.dstack((fred,fgreen,fblue))



#Laplacian
# def laplacianOperation(image):
#     red = image[:, :, 0]
#     fred = scipy.ndimage.convolve(laplace_sharpening_red, k, mode='constant', cval=0.0)

#     green = image[:, :, 1]
#     fgreen = scipy.ndimage.convolve(laplace_sharpening_green, k, mode='constant', cval=0.0)

#     blue = image[:, :, 2]
#     fblue = scipy.ndimage.convolve(laplace_sharpening_blue, k, mode='constant', cval=0.0)

#     return np.dstack((fred, fgreen, fblue))
```

```python
blurred_im = GaussianTreatment(noisy[1],3)
filter_blurred = GaussianTreatment(blurred_im,1)

alpha = 10
sharpened = blurred_im + alpha * (blurred_im - filter_blurred)

plt.figure(figsize=(18, 6))
plt.subplot(131)
plt.imshow(noisy[1])
plt.axis('off')
plt.subplot(132)
plt.imshow(blurred_im)
plt.axis('off')
plt.subplot(133)
plt.imshow(sharpened)
plt.show()
```

```python
from scipy import ndimage, misc
import matplotlib.pyplot as plt
result = medianTreatment(noisy[1],2)
result2 = medianTreatment(result,3)
fig = plt.figure(figsize=(18, 6))
plt.axis('off')# show the filtered result in grayscale
plt.subplot(131)
plt.imshow(noisy[1])
plt.axis('off')# left side
plt.subplot(132) 
plt.imshow(result)
plt.axis('off')
plt.subplot(133)
plt.imshow(result2)
plt.show()
```
