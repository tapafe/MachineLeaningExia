```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import os
from scipy import misc
from skimage.io import imread_collection, imsave
from skimage import *
import cv2

#creating a collection with the available images
blurry = imread_collection('Dataset/Blurry/*.jpg')
```

```python
from skimage.filters import unsharp_mask
     
def process_blurryOne(collection):  
    result = unsharp_mask(collection, radius=1, amount=1)
    return result
```

```python
def process_blurry(collection):  
    Df= pd.DataFrame(columns=['FileName','Blur_rate_before','Blur_rate_After'])
    for file in range(len(collection)):
        ImageO = cv2.Laplacian(collection[file], cv2.CV_64F).var()
        result = unsharp_mask(collection[file], radius=1, amount=1) 
        ImageT = cv2.Laplacian(result, cv2.CV_64F).var()
        Df=Df.append({'FileName':collection.files[file],'Blur_rate_before':ImageO,'Blur_rate_After':ImageT},ignore_index=True)
        imsave('datasets/projet/blurryEnd/' + collection.files[file][-7:-4] + '.jpg', img_as_ubyte(result))
    return Df
```

```python
process_blurry(blurry)

```
